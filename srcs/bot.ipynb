```python
browser = await start(headless=True, config=Config(
    browser_args=["-enable-unsafe-extension-debugging"]
))

tab = await browser.get('https://deviceandbrowserinfo.com/are_you_a_bot')
await tab._prepare_headless()
await tab.get("https://chatgpt.com")
await tab.reload()
```

```python
await tab.reload()
cookie_bar_accept = await tab.find("Accept All", best_match=True, timeout=10)
if cookie_bar_accept:
    await cookie_bar_accept.click()

signup = await tab.find("Sign up", best_match=True, timeout=1)
if signup:
    await signup.click()
```
